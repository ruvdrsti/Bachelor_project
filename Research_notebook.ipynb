```python
from compChem.Hartree_Fock import Molecule
```

```python
import numpy as np
import psi4
```

```python
psi4.set_options({'basis': "STO-3G"})
distances = np.arange(0.2, 5, 0.1)
atom = Molecule("""
H
""")
atom.setGuess()
```

```python
E_H = atom.iterator()
```

```python

E_elek_list = []
E_total_list = []
for distance in distances:
    hydrogen = Molecule(
    f"""
    H 0 0 0
    H {distance} 0 0
    
    units au
    """)
    hydrogen.setGuess()
    E_total = hydrogen.iterator()
    E_total_list.append(E_total - 2*E_H)
```

```python
import matplotlib.pyplot as plt
plt.plot(distances, E_total_list)
plt.axis([0, 5, -0.2, 0.04])
```

```python
psi4.set_options({'basis': "cc-pvdz"})
h3 = Molecule("""
H 0 0 0
H 0 0.86602540378 0.5
H 0 0 1
units angstrom""")
h3.setGuess()
```
